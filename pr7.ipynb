```python
import os

os.makedirs("toolkit", exist_ok=True)
```

```python
with open("toolkit/file_operations.py", "w") as f:
    f.write("""
def create_file(name):
    open(name, 'w').close()
    print("File created!")

def write_file(name, data):
    with open(name, 'w') as f:
        f.write(data)
    print("Data written!")

def read_file(name):
    with open(name, 'r') as f:
        print(f.read())

def append_file(name, data):
    with open(name, 'a') as f:
        f.write(data)
    print("Data appended!")
""")
```

```python
with open("toolkit/math_utils.py", "w") as f:
    f.write("""
import math

def factorial(n):
    return math.factorial(n)

def compound_interest(p, r, t):
    return p * (1 + r/100) ** t

def area_circle(r):
    return math.pi * r * r
""")
```

```python
import datetime
import time
import random
import uuid
import math

from toolkit import file_operations
from toolkit import math_utils


def datetime_menu():
    print("\n1. Current Date Time\n2. Date Difference")
    ch = int(input("Enter: "))

    if ch == 1:
        print("Now:", datetime.datetime.now())
    elif ch == 2:
        d1 = input("Enter date1 (YYYY-MM-DD): ")
        d2 = input("Enter date2 (YYYY-MM-DD): ")
        d1 = datetime.datetime.strptime(d1, "%Y-%m-%d")
        d2 = datetime.datetime.strptime(d2, "%Y-%m-%d")
        print("Difference:", abs((d2 - d1).days), "days")


def math_menu():
    print("\n1. Factorial\n2. Compound Interest")
    ch = int(input("Enter: "))

    if ch == 1:
        n = int(input("Enter number: "))
        print("Factorial:", math_utils.factorial(n))
    elif ch == 2:
        p = float(input("Principal: "))
        r = float(input("Rate: "))
        t = float(input("Time: "))
        print("CI:", math_utils.compound_interest(p, r, t))


def random_menu():
    print("\n1. Random Number\n2. OTP")
    ch = int(input("Enter: "))

    if ch == 1:
        print("Random:", random.randint(1, 100))
    elif ch == 2:
        otp = random.randint(1000, 9999)
        print("OTP:", otp)


def uuid_menu():
    print("UUID:", uuid.uuid4())


def file_menu():
    print("\n1. Create\n2. Write\n3. Read\n4. Append")
    ch = int(input("Enter: "))

    name = input("File name: ")

    if ch == 1:
        file_operations.create_file(name)
    elif ch == 2:
        data = input("Enter data: ")
        file_operations.write_file(name, data)
    elif ch == 3:
        file_operations.read_file(name)
    elif ch == 4:
        data = input("Enter data: ")
        file_operations.append_file(name, data)


def explore_module():
    mod = input("Enter module name: ")
    module = __import__(mod)
    print(dir(module))


# ---------------- MAIN ----------------
while True:
    print("\n==== Multi Utility Toolkit ====")
    print("1. DateTime")
    print("2. Math")
    print("3. Random")
    print("4. UUID")
    print("5. File Ops")
    print("6. Explore Module")
    print("7. Exit")

    ch = int(input("Enter choice: "))

    if ch == 1:
        datetime_menu()
    elif ch == 2:
        math_menu()
    elif ch == 3:
        random_menu()
    elif ch == 4:
        uuid_menu()
    elif ch == 5:
        file_menu()
    elif ch == 6:
        explore_module()
    elif ch == 7:
        print("Goodbye 👋")
        break
```
