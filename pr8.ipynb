```python
import numpy as np

class DataAnalytics:
    def __init__(self):
        self.array = None


    def create_array(self):
        print("\n1. 1D Array\n2. 2D Array\n3. 3D Array")
        choice = int(input("Enter choice: "))

        if choice == 1:
            size = int(input("Enter number of elements: "))
            elements = list(map(int, input("Enter elements: ").split()))
            self.array = np.array(elements)

        elif choice == 2:
            r = int(input("Rows: "))
            c = int(input("Columns: "))
            elements = list(map(int, input("Enter elements: ").split()))
            self.array = np.array(elements).reshape(r, c)

        elif choice == 3:
            x = int(input("Depth: "))
            y = int(input("Rows: "))
            z = int(input("Columns: "))
            elements = list(map(int, input("Enter elements: ").split()))
            self.array = np.array(elements).reshape(x, y, z)

        print("\nArray Created:\n", self.array)


    def indexing_slicing(self):
        print("\n1. Indexing\n2. Slicing")
        choice = int(input("Enter choice: "))

        if choice == 1:
            index = int(input("Enter index: "))
            print("Value:", self.array[index])

        elif choice == 2:
            print("Enter slice (start:end)")
            s = input("Row slice: ")
            if len(self.array.shape) > 1:
                c = input("Column slice: ")
                print("Result:\n", self.array[eval(s), eval(c)])
            else:
                print("Result:\n", self.array[eval(s)])

    def math_operations(self):
        print("\n1. Add\n2. Subtract\n3. Multiply\n4. Divide")
        choice = int(input("Enter choice: "))

        elements = list(map(int, input("Enter elements of second array: ").split()))
        arr2 = np.array(elements).reshape(self.array.shape)

        if choice == 1:
            print("Result:\n", self.array + arr2)
        elif choice == 2:
            print("Result:\n", self.array - arr2)
        elif choice == 3:
            print("Result:\n", self.array * arr2)
        elif choice == 4:
            print("Result:\n", self.array / arr2)


    def combine_split(self):
        print("\n1. Combine\n2. Split")
        choice = int(input("Enter choice: "))

        if choice == 1:
            elements = list(map(int, input("Enter elements of second array: ").split()))
            arr2 = np.array(elements).reshape(self.array.shape)
            combined = np.vstack((self.array, arr2))
            print("Combined Array:\n", combined)

        elif choice == 2:
            parts = int(input("Enter number of splits: "))
            split_arr = np.array_split(self.array, parts)
            print("Split Arrays:")
            for i in split_arr:
                print(i)


    def search_sort_filter(self):
        print("\n1. Search\n2. Sort\n3. Filter")
        choice = int(input("Enter choice: "))

        if choice == 1:
            val = int(input("Enter value to search: "))
            result = np.where(self.array == val)
            print("Found at indices:", result)

        elif choice == 2:
            print("Sorted Array:\n", np.sort(self.array))

        elif choice == 3:
            val = int(input("Show values greater than: "))
            print("Filtered:\n", self.array[self.array > val])


    def aggregates(self):
        print("\n1. Sum\n2. Mean\n3. Median\n4. Std\n5. Variance")
        choice = int(input("Enter choice: "))

        if choice == 1:
            print("Sum:", np.sum(self.array))
        elif choice == 2:
            print("Mean:", np.mean(self.array))
        elif choice == 3:
            print("Median:", np.median(self.array))
        elif choice == 4:
            print("Std Dev:", np.std(self.array))
        elif choice == 5:
            print("Variance:", np.var(self.array))



def main():
    obj = DataAnalytics()

    while True:
        print("\n====== NumPy Analyzer ======")
        print("1. Create Array")
        print("2. Indexing / Slicing")
        print("3. Math Operations")
        print("4. Combine / Split")
        print("5. Search / Sort / Filter")
        print("6. Aggregates")
        print("7. Exit")

        choice = int(input("Enter choice: "))

        if choice == 1:
            obj.create_array()
        elif choice == 2:
            obj.indexing_slicing()
        elif choice == 3:
            obj.math_operations()
        elif choice == 4:
            obj.combine_split()
        elif choice == 5:
            obj.search_sort_filter()
        elif choice == 6:
            obj.aggregates()
        elif choice == 7:
            print("Thank you! 👋")
            break
        else:
            print("Invalid choice!")


main()
```
