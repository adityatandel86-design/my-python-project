```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

class SalesDataAnalyzer:

    def __init__(self):
        self.data = None
        self.last_plot = None

    def load_data(self):
        path = input("Enter CSV file path: ")
        try:
            self.data = pd.read_csv(path)
            print("✅ Data Loaded Successfully!")
        except:
            print("❌ Error loading file")

    def explore_data(self):
        if self.data is None:
            print("Load data first!")
            return

        print("\n1. Head\n2. Tail\n3. Columns\n4. Info\n5. Describe")
        choice = int(input("Enter choice: "))

        if choice == 1:
            print(self.data.head())
        elif choice == 2:
            print(self.data.tail())
        elif choice == 3:
            print(self.data.columns)
        elif choice == 4:
            print(self.data.info())
        elif choice == 5:
            print(self.data.describe())

    def clean_data(self):
        if self.data is None:
            return

        print("\n1. Show missing\n2. Fill mean\n3. Drop missing")
        choice = int(input("Enter choice: "))

        if choice == 1:
            print(self.data.isnull().sum())
        elif choice == 2:
            self.data.fillna(self.data.mean(numeric_only=True), inplace=True)
            print("Missing values filled!")
        elif choice == 3:
            self.data.dropna(inplace=True)
            print("Rows dropped!")

    def dataframe_operations(self):
        print("\n1. Search\n2. Sort\n3. Filter")
        choice = int(input("Enter choice: "))

        if choice == 1:
            col = input("Column name: ")
            val = input("Value: ")
            print(self.data[self.data[col] == val])

        elif choice == 2:
            col = input("Sort by column: ")
            print(self.data.sort_values(by=col))

        elif choice == 3:
            col = input("Column: ")
            val = float(input("Greater than value: "))
            print(self.data[self.data[col] > val])

    def aggregate_functions(self):
        print("\n1. Sum\n2. Mean\n3. Count")
        choice = int(input("Enter choice: "))

        if choice == 1:
            print(self.data.sum(numeric_only=True))
        elif choice == 2:
            print(self.data.mean(numeric_only=True))
        elif choice == 3:
            print(self.data.count())

    def statistical_analysis(self):
        print("\n1. Std\n2. Variance\n3. Percentile")
        choice = int(input("Enter choice: "))

        if choice == 1:
            print(self.data.std(numeric_only=True))
        elif choice == 2:
            print(self.data.var(numeric_only=True))
        elif choice == 3:
            print(self.data.quantile(0.5))

    def pivot_table(self):
        col1 = input("Index column: ")
        col2 = input("Values column: ")
        print(pd.pivot_table(self.data, values=col2, index=col1, aggfunc='mean'))

    def visualize_data(self):
        print("\n1. Bar\n2. Line\n3. Scatter\n4. Histogram\n5. Pie\n6. Heatmap")
        choice = int(input("Enter choice: "))

        if choice == 1:
            x = input("X column: ")
            y = input("Y column: ")
            self.data.plot(kind='bar', x=x, y=y)
            self.last_plot = plt

        elif choice == 2:
            x = input("X column: ")
            y = input("Y column: ")
            plt.plot(self.data[x], self.data[y])
            self.last_plot = plt

        elif choice == 3:
            x = input("X column: ")
            y = input("Y column: ")
            plt.scatter(self.data[x], self.data[y])
            self.last_plot = plt

        elif choice == 4:
            col = input("Column: ")
            plt.hist(self.data[col])
            self.last_plot = plt

        elif choice == 5:
            col = input("Column: ")
            self.data[col].value_counts().plot.pie(autopct='%1.1f%%')
            self.last_plot = plt

        elif choice == 6:
            sns.heatmap(self.data.corr(numeric_only=True), annot=True)
            self.last_plot = plt

        plt.title("Visualization")
        plt.show()

    def save_plot(self):
        if self.last_plot:
            name = input("Enter filename (e.g. plot.png): ")
            self.last_plot.savefig(name)
            print("✅ Plot Saved!")
        else:
            print("No plot available!")



def main():
    obj = SalesDataAnalyzer()

    while True:
        print("\n====== Pandas Analyzer ======")
        print("1. Load Dataset")
        print("2. Explore Data")
        print("3. DataFrame Operations")
        print("4. Handle Missing Data")
        print("5. Aggregate Functions")
        print("6. Statistical Analysis")
        print("7. Pivot Table")
        print("8. Visualization")
        print("9. Save Plot")
        print("10. Exit")

        choice = int(input("Enter choice: "))

        if choice == 1:
            obj.load_data()
        elif choice == 2:
            obj.explore_data()
        elif choice == 3:
            obj.dataframe_operations()
        elif choice == 4:
            obj.clean_data()
        elif choice == 5:
            obj.aggregate_functions()
        elif choice == 6:
            obj.statistical_analysis()
        elif choice == 7:
            obj.pivot_table()
        elif choice == 8:
            obj.visualize_data()
        elif choice == 9:
            obj.save_plot()
        elif choice == 10:
            print("Goodbye 👋")
            break
        else:
            print("Invalid choice!")

main()
```
