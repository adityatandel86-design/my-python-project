```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

data = {
    "Date": [
        "2024-01-01","2024-01-02","2024-01-03","2024-01-04","2024-01-05",
        "2024-01-06","2024-01-07","2024-01-08","2024-01-09","2024-01-10"
    ],
    "Product": [
        "Laptop","Mobile","Chair","Table","T-Shirt",
        "Jeans","Laptop","Mobile","Chair","Table"
    ],
    "Category": [
        "Electronics","Electronics","Furniture","Furniture","Clothing",
        "Clothing","Electronics","Electronics","Furniture","Furniture"
    ],
    "Price": [50000,20000,3000,7000,500,1500,50000,20000,3000,7000],
    "Quantity Sold": [2,3,4,2,10,5,1,2,3,1]
}

df = pd.DataFrame(data)
df["Total Sales"] = df["Price"] * df["Quantity Sold"]
class RetailAnalyzer:

    def __init__(self, data):
        self.data = data


    def validate_data(self):
        print("\n--- Input Validation ---")
        required = ["Date","Product","Category","Price","Quantity Sold","Total Sales"]
        for col in required:
            if col in self.data.columns:
                print(col, "")
            else:
                print(col, " Missing")


    def load_data(self):
        print("\n--- Data Loading ---")
        print(self.data.head())


    def handle_data(self):
        print("\n--- Data Handling ---")

        self.data["Date"] = pd.to_datetime(self.data["Date"])


        self.data = self.data.sort_values(by="Date")

        print(self.data.head())


    def manipulate_data(self):
        print("\n--- Data Manipulation ---")
        self.data["Discount"] = self.data["Total Sales"] * 0.1
        self.data["Net Sales"] = self.data["Total Sales"] - self.data["Discount"]
        print(self.data.head())


    def analysis_metrics(self):
        print("\n--- Analysis & Metrics ---")
        print("Total Sales:", self.data["Total Sales"].sum())
        print("Average Sales:", self.data["Total Sales"].mean())

        popular = self.data.groupby("Product")["Total Sales"].sum().idxmax()
        print("Most Popular Product:", popular)


    def filter_data(self):
        print("\n--- Filter (Sales > 10000) ---")
        print(self.data[self.data["Total Sales"] > 10000])


    def visualize(self):
        print("\n--- Visualization ---")


        self.data.groupby("Category")["Total Sales"].sum().plot(kind="bar")
        plt.title("Sales by Category")
        plt.show()


        trend = self.data.groupby("Date")["Total Sales"].sum()
        trend.plot()
        plt.title("Sales Trend")
        plt.show()


        sns.heatmap(self.data.corr(numeric_only=True), annot=True)
        plt.title("Heatmap")
        plt.show()



obj = RetailAnalyzer(df)

obj.validate_data()
obj.load_data()
obj.handle_data()
obj.manipulate_data()
obj.analysis_metrics()
obj.filter_data()
obj.visualize()
```
